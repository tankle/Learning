```python
import tensorflow as tf
import numpy as np
```

```python
# 随机生成x的值
t_x = np.floor(10 * np.random.random([5]),dtype=np.float32)
```

```python
t_x
```

```python
# 确定一个方程，根据已有的x值生成对应的y值
t_y = t_x * 4.0 + 2.0
t_y
```

```python
# 训练集已经生成好，下面开始用Tensorflow 来推导出 参数a = 4, b = 2
x = tf.placeholder(tf.float32)
y = tf.placeholder(tf.float32)
```

```python
# 初始化两个变量
a = tf.Variable(0.0)
b = tf.Variable(0.0)
```

```python
# 生成一个方程描述式
now_y = x * a + b
```

```python
# 为了能够让机器知道自己当前学到的值是不是对的，就需要和正确的值做一个对比，叫做损失函数\
# 这里选择了平方误差：每个训练结果的差值的平方和
loss = tf.reduce_sum(tf.square(now_y - y))
```

```python
# 梯度下降
optimizer = tf.train.GradientDescentOptimizer(0.001)
train = optimizer.minimize(loss)
```

```python
# tensorflow 是一种图计算模型，启动运算需要创建一个session
session = tf.Session()
# 初始全局变量
session.run(tf.global_variables_initializer())
```

```python
for i in range(10000):
    session.run(train, {x: t_x, y:t_y})
    print(session.run([a, b, loss], {x:t_x, y:t_y}))
```
